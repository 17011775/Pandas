```
import pandas as pd 
```

```
file_url = 'https://bit.ly/dsa-01-order'
order = pd.read_csv(file_url)
order.set_index('id',inplace = True)
order.head()
```

##정렬

```
order.sort_index()
```

```
order.sort_index(ascending= False) #내림차순
```

```
order.sort_values(by = 'price') #컬럼 값 정렬
```

```
order.sort_values(by = 'price',ascending= False) 
```

```
order.sort_index(axis = 1) #axis = 1은 컬럼, 0은 로우
#order.sort_index(axis = 'columns')
```

```
order.sort_index(axis = 1,ascending= False)
```

```
order.sort_values(by = ['date','price']) #date 먼저 정렬, date 같으면 price 보고
```

```
order.sort_values(by = ['date','price'], ascending= [True, False]) #date는 오름차순, date가 같으면 price 내림차순
```

```
order['price'].sort_values(ascending = False).head()
```

##Group by

*  pivot_table : 통계연산 결과를 눈으로 확인
*  groupby : 통계연산 결과를 저장 또는 머신러닝 알고리즘에 넣을 때 



```
order.groupby('state')['price'].mean()
```

```
order.groupby(['date','state'])['price'].mean()
```

##여러개의 데이터를 하나로 합치기

```
url = 'https://bit.ly/dsa-02-order-201701'
order201701 = pd.read_csv(url, index_col= 'id')
order201701
```

```
url = 'https://bit.ly/dsa-02-order-201702'
order201702 = pd.read_csv(url, index_col= 'id')
order201702
```

```
order = pd.concat([order201701,order201702])
order
```

```
url = 'https://bit.ly/dsa-02-order-how'
how = pd.read_csv(url, index_col= 'id')
how
```

```
pd.concat([order,how], axis = 'columns')
```

```
url = 'https://bit.ly/dsa-02-user'
user = pd.read_csv(url)
user
```

```
url = 'https://bit.ly/dsa-02-product'
product = pd.read_csv(url)
product
```

```
order
```

```
pd.merge(order,user, left_on = 'user_id', right_on = 'id')
```

```
user['user_id'] = user.index
pd.merge(order,user,on = 'user_id')
```

```
pd.merge(order,user)
```

```
pd.merge(order,product,left_on = 'user_id', right_on = 'id')
```

```
pd.merge(order,product,left_on = 'product_id', right_on = 'id')
```

```
pd.merge(order,user,on = 'user_id', how = 'inner') #inner 교집합
```

```
pd.merge(order,user,on = 'user_id', how = 'outer') #outer 합집합
```

```
pd.merge(order,user,on = 'user_id', how = 'left') #left 왼쪽살리고 오른쪽버리기
# 메인데이터프레임을 왼쪽에 두고 left merge 하면 대부분 잘나온다
```

```
pd.merge(order,user,on = 'user_id', how = 'right')
```
