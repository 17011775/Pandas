##컬럼 추가&수정

```
data['지역'] = '서울'
data 
```

```
data['결제'] = [True, True, True, False, True, True, False, True, True,]
data
```

```
data['몸무게(gram)'] = data['몸무게'] * 1000
data
```

```
data['과체중'] = data['몸무게'] > 75.8
data
```

```
data.loc[data['몸무게'] > 75.8, '상태'] = '과체중'
data.loc[data['몸무게'] <= 75.8, '상태'] = '정상'
data
```

##컬럼 삭제

```
data = data.drop(['지역', '결제'], axis = 'columns')
data
```
